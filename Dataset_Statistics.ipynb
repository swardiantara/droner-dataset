```python
train_group = train_df.groupby(['sentence_id'], as_index=False)['words', 'labels'].agg(lambda x: list(x))
train_group
```

```python
length_list = []
for i in range(0, train_group.shape[0]):
    word_length = train_group.loc[i, 'words']
    length_list.append(len(word_length))
```

```python
train_group["length"] = length_list
train_group
```

```python
avg = sum(length_list)/len(length_list)
print(avg)
```

```python
train_group.length.describe()
```

```python
test_group = test_df.groupby(['sentence_id'], as_index=False)['words', 'labels'].agg(lambda x: list(x))
```

```python
length_list = []
for i in range(0, test_group.shape[0]):
    word_length = test_group.loc[i, 'words']
    length_list.append(len(word_length))
```

```python
test_group["length"] = length_list
test_group
```

```python
test_group.length.describe()
```

```python
full_df = pd.concat([train_group, test_group])
full_df.shape
```

```python
length_distribution = full_df.length.value_counts().to_frame()
length_distribution
```

```python
length_distribution = length_distribution.sort_index()
```

```python
length_distribution.describe()
```

```python
less_than_ten = []
for i in range(0, length_distribution.shape[0]):
    if(i < 10):
        less_than_ten.append(length_distribution.iloc[i])
```

```python
less_than_ten = pd.DataFrame(less_than_ten)
less_than_ten
```
