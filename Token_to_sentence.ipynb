```python
import spacy
import json
import pandas as pd
```

```python
tokenized = pd.read_csv('hasil_tools_197.csv', encoding='utf8')
```

```python
tokenized = tokenized.iloc[: , 1:]
tokenized
```

```python
tokenized = tokenized[tokenized.sentence_idx != 'O'] # Remove sentence separator
```

```python
tokenized = tokenized.dropna() # Drop nan values
```

```python
# Groupby and collect columns
grouped = tokenized.groupby(['sentence_idx'], as_index=False)['token', 'label'].agg(lambda x: list(x))
grouped
```

```python
grouped.to_csv('unique_sentence_to_label.csv')
```
